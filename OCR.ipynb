```
import cv2
import pytesseract
import matplotlib.pyplot as plt
from skimage import io
%matplotlib inline
```

```
image_1 = io.imread('https://upload.wikimedia.org/wikipedia/commons/thumb/b/bc/Wikinews_Breaking_News.png/800px-Wikinews_Breaking_News.png')
io.imshow(image_1)
```

```
text_1 = pytesseract.image_to_string(image_1)
print(text_1)
```
